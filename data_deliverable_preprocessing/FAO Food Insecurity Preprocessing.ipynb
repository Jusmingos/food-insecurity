## Getting all the values for a year

```python
food_df = food_df[["Area", "Year", "Item", "Value"]]
food_df.head(-5)
```

```python
# Edit the year column to display a single year - (starting year for the 3-year average columns)
for i in range(food_df.shape[0]):
  value = food_df.at[i, "Year"]
  splitted_years = value.split("-")
  # if year is like 2000-2001, split, otheriwse leave alone
  if len(splitted_years) == 2:
    food_df.loc[i, "Year"] = str(int(splitted_years[0]))

food_df.head(5)
```

```python
# Get columns
print(food_df['Item'].value_counts())
```

```python
# Combine all the entries for each country, year into one row
new_df = pd.DataFrame() 
for i in range(food_df.shape[0]):
  # Index by country,year for joining purposes
  index = food_df.at[i, "Area"] + "," + food_df.at[i, "Year"]
  # Get entry 
  col = food_df.at[i, "Item"]
  value = food_df.at[i, "Value"]
  # Make country, year columns in addition to index
  new_df.at[index, "Country"] = food_df.at[i, "Area"]
  new_df.at[index, "Year"] = food_df.at[i, "Year"]
  # Get value for column 
  new_df.at[index, col] = value
new_df.head()
```

```python
new_df['Prevalence of severe food insecurity in the total population (percent) (3-year average)'].describe()
```

```python
# Drop cols without a lot of data
drop_cols = [] 
for col in new_df.columns:
  # Majority of data is null, remove
  if new_df[col].isna().sum() >= 2200:
    drop_cols.append(col)

food_data = new_df.drop(drop_cols, axis=1)
food_data
```
